```python
import pandas as pd

df = pd.read_csv('./archive/imdb_top_1000.csv')

# Display the first few rows
df.head()
```

1. What are the column names in the dataset?

```python
column_names = df.columns
print(column_names.tolist())
```

2. What is the average rating of the movies?

```python
average_rating=df['IMDB_Rating'].mean()
print(average_rating)
```

3. How many unique genres are there?

```python
unique_genres = df['Genre'].nunique()
print(unique_genres)
```

4.Save the Results as a CSV File

```python
# Create a DataFrame with the results
results = pd.DataFrame({'Question': ['What are the column names in the dataset?',
                                     'What is the average rating of the movies?',
                                     'How many unique genres are there?'],
                       'Answer': [', '.join(column_names),
                                  str(average_rating),
                                  str(unique_genres)]})

results.to_csv('./result.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Calculate the genre distribution
genre_counts = df['Genre'].value_counts()

# genre distribution
plt.figure(figsize=(12, 8))
plt.bar(genre_counts.index, genre_counts.values, color='#FF7F0E')  # Choose a fancy color for the bars
plt.xlabel('Genre', fontsize=14)
plt.ylabel('Number of Movies', fontsize=14)
plt.title('Genre Distribution of Movies', fontsize=16, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)

# Add a fancy background color to the plot
plt.gca().set_facecolor('#F8F8F8')
plt.grid(axis='y', color='white', linestyle='-', linewidth=0.5)

# Add a fancy border around the plot
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Show the plot
plt.show()


# Calculate the number of movies per year
year_counts = df['Released_Year'].value_counts().sort_index()

# Create a fancy scatter plot for number of movies vs year
plt.figure(figsize=(12, 8))
plt.scatter(year_counts.index, year_counts.values, s=100, color='#1F77B4')  # Choose a fancy color for the scatter plot
plt.xlabel('Year', fontsize=14)
plt.ylabel('Number of Movies', fontsize=14)
plt.title('Number of Movies vs Year', fontsize=16, fontweight='bold')
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Add a fancy background color to the plot
plt.gca().set_facecolor('#F8F8F7')
plt.grid(color='white', linestyle='-', linewidth=0.5)

# Add a fancy border around the plot
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# print the plot
plt.show()


# lets print the number of movies from each genre
print("no of movie from each genre:")
print(genre_counts)
```

```python
import seaborn as sns
year_counts = df['Released_Year'].value_counts().sort_index()

# Set up a fancy plot style using Seaborn
sns.set(style='darkgrid')
sns.set(font='Arial')

# Create a figure and axis with fancy dimensions
fig, ax = plt.subplots(figsize=(12, 8))

# Customize the scatter plot with a fancy design
ax.scatter(year_counts.index, year_counts.values, s=100, c='#FF7F0E', alpha=0.7)  # Choose a fancy color for the scatter plot
ax.set_xlabel('Year', fontsize=14, fontweight='bold')
ax.set_ylabel('Number of Movies', fontsize=14, fontweight='bold')
ax.set_title('Number of Movies vs Year', fontsize=16, fontweight='bold')

# Set the tick label font size and weight
ax.tick_params(axis='both', labelsize=12, width=2)
for tick in ax.get_xticklabels():
    tick.set_fontweight('bold')
for tick in ax.get_yticklabels():
    tick.set_fontweight('bold')


ax.set_facecolor('#F8F8F8')

# Add a fancy border around the plot
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_linewidth(1.5)
ax.spines['left'].set_linewidth(1.5)

# Show the plot
plt.show()
```
